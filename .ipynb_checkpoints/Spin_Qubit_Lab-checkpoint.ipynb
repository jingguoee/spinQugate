Spin Quantum Gate Lab is a set of python codes that calculate the device-level characteristics of spin-based quantum gates. 
Single-qubit rotational gates and two-qubit controlled-phase gate can be simulated, which form a complete set of quantum gates for universal quantum computing.
The tool can simulate time evolution, compute delay and fidelity, and perform quantum process tomography for the quantum gates. 
The effect of dephasing can be simulated by inputting a spin dephasing time.

 
The simulation modules are listed below:

For simulation of "1. single qubit rotation gate", a magnetic field is applied to rotate the spin of a qubit in the presence of dephasing. 

[Single_Quantum_Gate](./Single_Quantum_Gate.ipynb)

For simulation of "2. Two qubit CZ gate", a double quantum dot (QD) structure is simulated by applying a detuning gate voltage in the presence of dephasing.  

[Double_Quantum_Gate](./Double_Quantum_Gate.ipynb)

Examples of the simulation are provided in supporting docs.
